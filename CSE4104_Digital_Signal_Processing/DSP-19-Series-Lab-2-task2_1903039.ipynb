```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
```

## 1. Convolution operation using Gaussian impulse/kernel/filter:

- Convolution is a mathematical operation that combines two functions to produce a third function that expresses how the shape of one function is modified by the other.
- When performing convolution in signal processing, we often use an impulse response (also known as a kernel or filter) to modify our input signal.

- The Gaussian function is given by:
- $$ G(x,y) = ( 1/{(2*π*σ^2) } )*e ^ {- (x^2 + y^2)/ 2*σ^2 } $$
- σ is the standard deviation.


### 1.1 Define Gaussian Kernel


```python
def gaussian_kernel(size, sigma=1):
    """Generates a Gaussian kernel."""
    x = np.linspace(-size // 2, size // 2, size)
    g = np.exp(-0.5 * (x / sigma)**2)
    return g / g.sum()
```

### 1.2 Convolve signal with Gaussian kernal


```python
def convolve_signal_with_gaussian(input_signal, kernel_size, sigma=1):
    """Convolves a signal with a Gaussian kernel."""
    kernel = gaussian_kernel(kernel_size, sigma)
    convolved_signal = signal.convolve(input_signal, kernel, mode='same')
    return convolved_signal
```

### 1.3 Create a signal & Convulate it with Gaussian Filter


```python
input_signal = np.sin(np.linspace(0, 2 * np.pi, 100))
kernel_size = 5
sigma = 1.0

convolved_signal = convolve_signal_with_gaussian(input_signal, kernel_size, sigma)
```

### 1.4 Plotting the original and convolved signals


```python
plt.figure(figsize=(15, 7))
plt.plot(input_signal, label='Original Signal')
plt.plot(convolved_signal, label='Convolved Signal', linestyle='dashed')
plt.legend()
plt.title('Signal Convolution with Gaussian Kernel')
plt.show()
```

# 2. Correlation of signals

The measure of similarity between two signals, functions, or waveforms is called correlation.


### 2.1 Define function for Correration of two signals


```python
def correlate_signals(signal1, signal2):
    """Performs correlation between two signals."""
    correlation = np.correlate(signal1, signal2, mode='full')
    return correlation
```

### 2.2 Create 2 signals & Correlate them


```python
signal1 = np.sin(np.linspace(0, 2 * np.pi, 100))
signal2 = np.sin(np.linspace(0, 2 * np.pi, 100) + np.pi / 4)  # Phase-shifted version of signal1

correlation = correlate_signals(signal1, signal2)
```

### 2.3 Plotting the original signals and their correlation


```python
plt.figure(figsize=(15, 7))
plt.subplot(2, 1, 1)
plt.plot(signal1, label='Signal 1')
plt.plot(signal2, label='Signal 2', linestyle='dashed')
plt.legend()
plt.title('Original Signals')

plt.subplot(2, 1, 2)
plt.plot(correlation, label='Correlation')
plt.legend()
plt.title('Correlation of Signals')

plt.tight_layout()
plt.show()
```
