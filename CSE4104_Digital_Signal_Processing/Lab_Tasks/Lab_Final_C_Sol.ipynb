# Q1: Drawing Discrete Time Signals

## Objective:
In this lab, we will draw the following discrete time signals over the time range from -2 to +2.5 with an increment of 0.01:

1. **Exponential Signal** (Basic Form: \( e^n \))
2. **Unit Ramp**
3. **Unit Sample**
4. **Unit Step**
5. **Sinusoidal Signals** (Both Sine & Cosine)

We will set the time period \( T = 0.125 \) seconds and the phase \( &theta; = 0&deg; \) for the sinusoidal signals.


````python
# Q1_Signals.py

import numpy as np
import matplotlib.pyplot as plt

# Define the time vector from -2 to +2.5 with an increment of 0.01
n = np.arange(-2, 2.51, 0.01)

# a. Exponential Signal: e^n
exp_signal = np.exp(n)

# b. Unit Ramp: r[n] = n for n >= 0, else 0
ramp_signal = np.where(n >= 0, n, 0)

# c. Unit Sample: δ[n], 1 at n=0, else 0
unit_sample = np.where(n == 0, 1, 0)

# d. Unit Step: u[n], 1 for n >= 0, else 0
unit_step = np.where(n >= 0, 1, 0)

# e. Sinusoidal Signals
# Given: Time period T = 0.125 seconds, Phase = 0 degrees
T = 0.125  # Time period in seconds
f = 1 / T  # Frequency
omega = 2 * np.pi * f  # Angular frequency

phase_deg = 0  # Phase in degrees
phase_rad = np.deg2rad(phase_deg)  # Convert phase to radians

sin_signal = np.sin(omega * n + phase_rad)
cos_signal = np.cos(omega * n + phase_rad)

# Plotting the signals
plt.figure(figsize=(14, 10))

# a. Exponential Signal
plt.subplot(3, 2, 1)
plt.plot(n, exp_signal, label='$e^n$')
plt.title('Exponential Signal $e^n$')
plt.xlabel('n')
plt.ylabel('$e^n$')
plt.grid(True)
plt.legend()

# b. Unit Ramp
plt.subplot(3, 2, 2)
plt.plot(n, ramp_signal, label='Unit Ramp r[n]', color='orange')
plt.title('Unit Ramp Signal $r[n]$')
plt.xlabel('n')
plt.ylabel('$r[n]$')
plt.grid(True)
plt.legend()

# c. Unit Sample
plt.subplot(3, 2, 3)
plt.plot(n, unit_sample, label='Unit Sample δ[n]', color='green')
plt.title('Unit Sample Signal $\delta[n]$')
plt.xlabel('n')
plt.ylabel('$\delta[n]$')
plt.grid(True)
plt.legend()

# d. Unit Step
plt.subplot(3, 2, 4)
plt.plot(n, unit_step, label='Unit Step u[n]', color='red')
plt.title('Unit Step Signal $u[n]$')
plt.xlabel('n')
plt.ylabel('$u[n]$')
plt.grid(True)
plt.legend()

# e. Sinusoidal Signals
plt.subplot(3, 2, 5)

plt.plot(n, sin_signal, label='Sine Signal', color='blue')
plt.plot(n, cos_signal, label='Cosine Signal', color='magenta', linestyle='--')

plt.title('Sinusoidal Signals')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

## Explanation:

### Time Vector (n):
- Defined from \( -2 \) to \( +2.5 \) with an increment of 0.01 using `numpy.arange`.

### Signals:
- **Exponential Signal \( e^n \)**: Computed using `numpy.exp(n)`.
- **Unit Ramp \( r[n] \)**: Defined as \( n \) when \( n &geq; 0 \), else 0.
- **Unit Sample \( &delta;[n] \)**: Equals 1 at \( n = 0 \), else 0.
- **Unit Step \( u[n] \)**: Equals 1 for \( n &geq; 0 \), else 0.

### Sinusoidal Signals:
- **Frequency \( f \)**:  \( $\frac{1}{T}$ \) , where \( T \) is the time period.
- **Angular Frequency \( &omega; \)**: \( 2&pi; f \).
- **Phase**: Converted from degrees to radians.
- **Sine and Cosine**: Computed using `numpy.sin` and `numpy.cos`.

### Plotting:
- Each signal is plotted in a separate subplot for clarity.
- Titles, labels, grids, and legends are added for better visualization.

### Output:
- Running this script will display a window with five subplots, each showcasing one of the discrete time signals as specified.

## Q2: Using Stem Function to Observe Signals from Q1

### Objective:
- Use the `stem` function to visualize the discrete time signals from Q1.


````python
# Q2_StemPlots.py

import numpy as np
import matplotlib.pyplot as plt

# Define the time vector from -2 to +2.5 with an increment of 1 (for discrete time)
n = np.arange(-2, 3, 1)  # n = -2, -1, 0, 1, 2

# a. Exponential Signal: e^n
exp_signal = np.exp(n)

# b. Unit Ramp: r[n] = n for n >= 0, else 0
ramp_signal = np.where(n >= 0, n, 0)

# c. Unit Sample: δ[n], 1 at n=0, else 0
unit_sample = np.where(n == 0, 1, 0)

# d. Unit Step: u[n], 1 for n >= 0, else 0
unit_step = np.where(n >= 0, 1, 0)

# e. Sinusoidal Signals
T = 0.125  # Time period in seconds
f = 1 / T
omega = 2 * np.pi * f
phase_deg = 0
phase_rad = np.deg2rad(phase_deg)

sin_signal = np.sin(omega * n + phase_rad)
cos_signal = np.cos(omega * n + phase_rad)

# Plotting the signals using stem
plt.figure(figsize=(14, 10))

# a. Exponential Signal
plt.subplot(3, 2, 1)
plt.stem(n, exp_signal, basefmt=" ", use_line_collection=True)
plt.title('Exponential Signal $e^n$')
plt.xlabel('n')
plt.ylabel('$e^n$')
plt.grid(True)

# b. Unit Ramp
plt.subplot(3, 2, 2)
plt.stem(n, ramp_signal, basefmt=" ", linefmt='C1-', markerfmt='C1o', use_line_collection=True)
plt.title('Unit Ramp Signal $r[n]$')
plt.xlabel('n')
plt.ylabel('$r[n]$')
plt.grid(True)

# c. Unit Sample
plt.subplot(3, 2, 3)
plt.stem(n, unit_sample, basefmt=" ", linefmt='C2-', markerfmt='C2o', use_line_collection=True)
plt.title('Unit Sample Signal $\delta[n]$')
plt.xlabel('n')
plt.ylabel('$\delta[n]$')
plt.grid(True)

# d. Unit Step
plt.subplot(3, 2, 4)
plt.stem(n, unit_step, basefmt=" ", linefmt='C3-', markerfmt='C3o', use_line_collection=True)
plt.title('Unit Step Signal $u[n]$')
plt.xlabel('n')
plt.ylabel('$u[n]$')
plt.grid(True)

# e. Sinusoidal Signals
plt.subplot(3, 2, 5)
plt.stem(n, sin_signal, linefmt='C4-', markerfmt='C4o', basefmt=" ", label='Sine', use_line_collection=True)
plt.stem(n, cos_signal, linefmt='C5--', markerfmt='C5s', basefmt=" ", label='Cosine', use_line_collection=True)
plt.title('Sinusoidal Signals')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

## Explanation:

### Time Vector (n):
- Defined as integer values from \( -2 \) to \( +2 \) with an increment of 1, suitable for discrete time signals.

### Signals:
- Same as in Q1, but adjusted for discrete time where \( n \) is represented as integers.

### Plotting with Stem:
- The `stem` function is used to create discrete plots.
- Each signal is plotted in a separate subplot.
- Different colors and markers are used for clarity.
- Grids are added for better visualization.

### Output:
- Executing this script will display a window with five subplots, each illustrating one of the discrete time signals using stem plots.

## Q3: Shifting the Exponential Signal by 0.5 Units

### Objective:
- Shift the exponential signal from Q1-(a) by 0.5 units to both the left and right without using loops.
- Visualize the original and shifted signals using both `stem` and `plot` functions.

````python
# Q3_ShiftExponential.py

import numpy as np
import matplotlib.pyplot as plt

# Define the time vector from -2 to +2.5 with an increment of 0.01
n = np.arange(-2, 2.51, 0.01)

# Original Exponential Signal: e^n
exp_signal = np.exp(n)

# Shifting the signal by 0.5 units to the right: n - 0.5
n_shift_right = n - 0.5
exp_shift_right = np.exp(n_shift_right)

# Shifting the signal by 0.5 units to the left: n + 0.5
n_shift_left = n + 0.5
exp_shift_left = np.exp(n_shift_left)

# Plotting using plot function
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(n, exp_signal, label='Original', color='blue')
plt.plot(n, exp_shift_right, label='Shift Right by 0.5', color='red', linestyle='--')
plt.plot(n, exp_shift_left, label='Shift Left by 0.5', color='green', linestyle='-.')
plt.title('Exponential Signal Shifts (Plot)')
plt.xlabel('n')
plt.ylabel('$e^n$')
plt.legend()
plt.grid(True)

# Plotting using stem function
plt.subplot(1, 2, 2)
# For stem plots, select integer n for better visualization
n_discrete = np.arange(-2, 3, 1)
exp_signal_discrete = np.exp(n_discrete)

n_shift_right_discrete = n_discrete - 0.5
exp_shift_right_discrete = np.exp(n_shift_right_discrete)

n_shift_left_discrete = n_discrete + 0.5
exp_shift_left_discrete = np.exp(n_shift_left_discrete)

plt.stem(n_discrete, exp_signal_discrete, linefmt='C0-', markerfmt='C0o', basefmt=" ", label='Original', use_line_collection=True)
plt.stem(n_shift_right_discrete, exp_shift_right_discrete, linefmt='C1--', markerfmt='C1s', basefmt=" ", label='Shift Right by 0.5', use_line_collection=True)
plt.stem(n_shift_left_discrete, exp_shift_left_discrete, linefmt='C2-.', markerfmt='C2^', basefmt=" ", label='Shift Left by 0.5', use_line_collection=True)
plt.title('Exponential Signal Shifts (Stem)')
plt.xlabel('n')
plt.ylabel('$e^n$')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

## Explanation:

### Time Vector (n):
- Continuous time vector from \( -2 \) to \( +2.5 \) with a small increment (0.01) for smooth plotting.

### Shifting Without Loops:
- **Shift Right by 0.5**: Replace \( n \) with \( n - 0.5 \) in the exponential function.
- **Shift Left by 0.5**: Replace \( n \) with \( n + 0.5 \) in the exponential function.
- No loops are used; vectorized operations handle the shifting efficiently.

### Plotting:
- **Left Subplot (plot)**: Continuous plots showing the original and shifted signals with different styles.
- **Right Subplot (stem)**: Discrete stem plots for integer values of \( n \), showing the original and shifted signals with different markers and line styles.

### Output:
- Running this script will display a window with two subplots:
  - **Left**: Continuous line plots of the original and shifted exponential signals.
  - **Right**: Stem plots illustrating the discrete nature of the original and shifted signals.

## Q4: Downsampling the Exponential Signal

### Objective:
- Perform downsampling on the exponential signal from Q1-(a) with a scaling factor of 2.
- Plot both the original and downsampled signals in the same figure using subplots or overlays.

````python
# Q4_Downsampling.py

import numpy as np
import matplotlib.pyplot as plt

# Define the original time vector from -2 to +2.5 with an increment of 0.01
n = np.arange(-2, 2.51, 0.01)
exp_signal = np.exp(n)

# Downsampling with scaling factor 2
scaling_factor = 2
# Select every 2nd sample
n_down = n[::scaling_factor]
exp_down = exp_signal[::scaling_factor]

# Plotting
plt.figure(figsize=(12, 6))

# Original Signal
plt.subplot(2, 1, 1)
plt.plot(n, exp_signal, label='Original Signal', color='blue')
plt.title('Original Exponential Signal')
plt.xlabel('n')
plt.ylabel('$e^n$')
plt.grid(True)
plt.legend()

# Downsampled Signal
plt.subplot(2, 1, 2)
plt.stem(n_down, exp_down, linefmt='C1-', markerfmt='C1o', basefmt=" ", use_line_collection=True, label='Downsampled Signal (Factor=2)')
plt.title('Downsampled Exponential Signal')
plt.xlabel('n')
plt.ylabel('$e^n$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

## Explanation:

### Original Time Vector (n) and Signal (exp_signal):
- Defined as in previous questions.

### Downsampling Process:
- **Scaling Factor \( M \)**: Set to 2.
- **Downsampling Operation**: 
  - Select every \( M \)th sample from the original signal using slicing (e.g., `n[::2]`).

### Plotting:
- **Top Subplot**: 
  - Continuous plot of the original exponential signal.
- **Bottom Subplot**: 
  - Stem plot of the downsampled signal, clearly showing reduced sampling points.

### Note:
- Downsampling reduces the number of samples in the signal by the scaling factor. Here, with a scaling factor of 2, every second sample is selected.

### Output:
- Executing this script will display a window with two subplots:
  - **Top**: Continuous plot of the original exponential signal.
  - **Bottom**: Stem plot of the downsampled signal, highlighting the reduction in samples.
