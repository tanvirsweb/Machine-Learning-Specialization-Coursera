```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
plt.rcParams['figure.figsize'] = (14, 4)
```

```python
tf = 256
t = np.arange(0, 1, 1/tf)
f = 4
x = np.sin(2*np.pi*f*t)
plt.plot(t, x)
```

```python
SRATE = 16 * f
n = np.arange(0,1,1/SRATE)
nx = np.sin(2*np.pi*f*n)
plt.plot(t,x)
plt.stem(n, nx)
```

```python
import matplotlib.colors as mcolors

colors = list(mcolors.CSS4_COLORS.keys())
print(colors)
```

```python
plt.figure(figsize=(16,8))
qL = 8
ty = np.linspace(-1,1,qL)
plt.yticks(list(ty))
plt.plot(t,x)
plt.stem(n, nx)
for i in ty:
    plt.axhline(y=i, xmin=0, xmax=1, color='darkorange')
plt.show()
```

```python
xq = ty[np.abs(nx[:, None] - ty).argmin(axis=1)]
plt.figure(figsize=(16,8))
plt.yticks(list(ty))
plt.plot(t,x, color='aqua', label='original signal')
m, s, b = plt.stem(n, xq, label='quantized points')
plt.setp(m, color='brown')
plt.setp(s, color='brown')
plt.setp(b, color='black')

m, s, b = plt.stem(n, nx, label='sampled points', basefmt=None)
plt.setp(m, color='chartreuse')
plt.setp(s, visible=False)
plt.setp(b, color='black')
for i in ty:
    plt.axhline(y=i, xmin=0, xmax=1, color='darkorange')
plt.legend(fontsize=12)
plt.show()
```
