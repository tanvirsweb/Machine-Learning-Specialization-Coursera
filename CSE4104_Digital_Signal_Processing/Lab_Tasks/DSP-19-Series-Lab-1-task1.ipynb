```python
def plotSignal(n, x ,plot_title='Signal x(n) = sin(0.2πn)', xticks=(np.arange(-24.0, 24.0+1, 1)), yticks=(np.arange(-2.0, 2.0, 0.1))):
  plt.figure(figsize=(22, 10))
  plt.stem(n, x, 'b', markerfmt='bo', basefmt='r-', label='x(n) = cos(0.2πn)')

  plt.title( plot_title )
  plt.xlabel('n')
  plt.ylabel('x(n)')
  plt.xticks( xticks )  # x ticks from -20 to 20 with a step of 2
  plt.yticks( yticks )  # y ticks from -1.0 to 1.0 with a step of 0.2

  plt.legend()
  plt.grid()
  plt.show()
```

<h2>3. Signal Operations:</h2>




```python
def shift(signal, n, k):
    #  Shift the signal by k units
    # signal = signal + [0]*k
    return signal, n + k

def flip(signal, n):
    #  Flip the signal
    return signal[::-1], -n[::-1]

def scaleAmplitude(signal, n, a):
    #  Scale the signal
    return a * signal, n

# Function to scale time by a given factor
def scaleTime(signal, time_indices, factor):
    scaled_indices = time_indices[::factor]
    scaled_signal = signal[::factor]
    return scaled_signal, scaled_indices
    # time_indices[::factor] slices the time_indices array starting from the beginning, taking every factor-th element. For example, if factor is 2, it will take every second element from time_indices.
```
