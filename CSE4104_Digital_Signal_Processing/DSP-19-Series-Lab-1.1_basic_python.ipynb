```python
def factorial(n):
  if n==1:
    return 1;
  else:
    return n*factorial(n-1);

while(True):
  n = int(input('Enter a number(0 to exit): '))
  if n==0:
    break;
  fact = factorial(n)
  print('Factorial of:', n, 'is:', fact)
```
