# ***Changing the color spaces***


*   A color space is a protocol for representing colors in a way that makes them easily reproducible. We know that grayscale images have single pixel values and color images contain 3 values for each pixel the intensities of the Red, Green and Blue channels.
*  OpenCV reads a given image in the **BGR** format by **default**. So, youâ€™ll need to change the color space of your image from BGR to RGB when reading images using OpenCV.
* For **color conversion**, we use the function **cv2.cvtColor(input image, flag)** where flag deter- mines the type of conversion.

1. For BGR ->Gray conversion we use the flags cv2.COLOR_BGR2GRAY.
2. For BGR -> RGB conversion we use the flags cv2.COLOR_BGR2RGB.
3. For BGR -> HSV, we use the flag cv2.COLOR_BGR2HSV.


**Convert to Gray Scale**

```python
#converting the image into grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#display the image
cv2_imshow(image_gray)
```

**cv2.cvtColor:** This is a function from the OpenCV (Open Source Computer Vision) library. It's used to convert an image from one color space to another.  
**image:** This is the input image that you want to convert. It's assumed that you have loaded an image using OpenCV before this line of code.  
  
**cv2.COLOR_BGR2HSV:** This is the conversion code specifying the color space **conversion** you want to perform. In this case, it converts the image from the BGR (Blue, Green, Red) color space to HSV **(Hue, Saturation, Value)** color space. HSV is often preferred in image processing tasks because it **separates intensity** information (value) from color information (hue and saturation), which can be useful for various operations like **color segmentation, thresholding,** etc.

```python
#converting the image in HSV color space
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) #cvt: ConVerT
#display the image
cv2_imshow(image_hsv)
plt.imshow(image_hsv)
```

<h3>Saturation:</h3> The saturation value tells us how much **quantity** of respective color must be added. A 100% saturation means that complete pure color is added, while a **0% saturation** means no color is added, resulting in **grayscale**.

<h3>Value:</h3> The value represents the brightness concerning the saturation of the color. the value 0 represents total black darkness, while the value 100 will mean a full brightness and depend on the saturation.

<h3>Advantages:</h3>
The advantage of HSV is that it generalizes how **humans perceive color**. Hence it is the most accurate depiction of how we feel colors on the computer screen. Also, the HSV color space separates the luma from the color information. this allows us to perform the operations mentioned above in the applications section, as the histogram equalization will only be required for intensity values.

```python
# Converting the image to HSV color space
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Display the image using cv2_imshow
cv2_imshow(cv2.cvtColor(image_hsv, cv2.COLOR_HSV2BGR))

# Display the image using plt.imshow
plt.imshow(cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB))
plt.show()
```

# ***Splitting & Merging Image Channels***
* The B,G,R channels of an image can be split into their individual planes when needed.
* Then, the individual channels can be merged back together to form a BGR image again.
* In OpenCV, an image is represented as a NumPy array of shape (height, width, num_channels), where num_channels corresponds to the number of color channels in the image. For a color image, the num_channels is typically 3, corresponding to the Red, Green, and Blue channels (RGB).
* Splits the image array into separate pixel arrays for each color channel (Red, Green, and Blue) using the split () function, and stores them in the r, g, and b variables.
* Merges the individual color channels into a single RGB image using the merge () function.

```python
#Splitting Image Channels
b, g, r = cv2.split(image)

print("b:\n",b)
print("g:\n",g)
print("r:\n",r)

rgb_image = cv2.merge((r, g, b))
num_red_pixels = cv2.countNonZero(r)
num_green_pixels = cv2.countNonZero(g)
num_blue_pixels = cv2.countNonZero(b)

# Display the total number of pixels for each color channel
print('\nTotal number of red pixels:', num_red_pixels)
print('Total number of green pixels:', num_green_pixels)
print('Total number of blue pixels:', num_blue_pixels)
```

```python
#Displaying different image channels
#creates a figure with a size of 12 inches width by 10 inches height and adds two subplots to it.
fig = plt.figure(figsize=(12,10))
# Creating subplot for top left corner of the grid
ax1 = fig.add_subplot(121)
# Creating subplot for top right corner of the grid
ax2 = fig.add_subplot(122)

ax1.imshow(image)
ax1.title.set_text('BGR Channeled Image')
ax2.imshow(rgb_image)
ax2.title.set_text('RGB Channeled Image')
```

# ***Seperating Image Channels and Displaying Using Color Map***



This code snippet is using Matplotlib, a popular plotting library in Python, to display the red, green, and blue channels of an image side by side. Here's a breakdown:

1. `fig, axs = plt.subplots(1, 3, figsize=(10, 5))`: This line creates a figure (`fig`) and an array of subplots (`axs`) with 1 row and 3 columns, effectively creating a layout for three plots side by side. The `figsize` parameter specifies the width and height of the figure in inches.

2. `axs[0].imshow(r, cmap='gray')`: This line displays the red channel (`r`) of the image in the first subplot (`axs[0]`). The `cmap='gray'` argument specifies that the colormap used for displaying the image is grayscale.

3. `axs[0].set_title('Red Channel')`: This line sets the title of the first subplot to 'Red Channel'.

4. `axs[1].imshow(g, cmap='gray')`: Similarly, this line displays the green channel (`g`) of the image in the second subplot (`axs[1]`) and sets the colormap to grayscale.

5. `axs[1].set_title('Green Channel')`: This line sets the title of the second subplot to 'Green Channel'.

6. `axs[2].imshow(b, cmap='gray')`: Likewise, this line displays the blue channel (`b`) of the image in the third subplot (`axs[2]`) and sets the colormap to grayscale.

7. `axs[2].set_title('Blue Channel')`: This line sets the title of the third subplot to 'Blue Channel'.

8. `plt.show()`: Finally, this command displays the entire plot with the three subplots side by side.

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 5))
# Display the red channel image in the first subplot
axs[0].imshow(r, cmap='gray')
axs[0].set_title('Red Channel')

# Display the green channel image in the second subplot
axs[1].imshow(g, cmap='gray')
axs[1].set_title('Green Channel')

# Display the blue channel image in the third subplot
axs[2].imshow(b, cmap='gray')
axs[2].set_title('Blue Channel')

# Show the plot
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 5))
# 10 inches wide and 5 inches tall

# Display the red channel image in the first subplot with a colorful colormap
axs[0].imshow(r, cmap='Reds')  # Example colormap: 'Reds'
axs[0].set_title('Red Channel')

# Display the green channel image in the second subplot with a colorful colormap
axs[1].imshow(g, cmap='Greens')  # Example colormap: 'Greens'
axs[1].set_title('Green Channel')

# Display the blue channel image in the third subplot with a colorful colormap
axs[2].imshow(b, cmap='Blues')  # Example colormap: 'Blues'
axs[2].set_title('Blue Channel')

# Show the plot
plt.show()
```

# ***Image Resize and Rotation***
