# **Edge Detection**

- Edge detection is an image processing technique for finding the boundaries of objects within images.
- It works by detecting discontinuities in brightness.
- Edge detection is used for image segmentation.


```python
# Importing Required Libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
image = cv2.imread('img/road.bmp', cv2.IMREAD_GRAYSCALE)

fig = plt.figure(figsize=(15, 4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.imshow(image, cmap='gray')
ax1.title.set_text('Original Image')
ax2.hist(image.ravel(), bins=256)
ax2.title.set_text('Original Image Distribution')
```

# Laplacian Edge Detection


```python
# Importing the Libraries & Loading the Image
import cv2
import matplotlib.pyplot as plt
import numpy as np
image = cv2.imread('img/road.bmp', cv2.IMREAD_GRAYSCALE)
# Laplace Filter
kernel = np.array([[0, 1, 0],
                   [1,-4, 1],
                   [0, 1, 0]])

laplacian = cv2.filter2D(image, -1, kernel)

fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.imshow(image, cmap='gray', vmin=0, vmax=255)
ax1.title.set_text('Original Image')
ax2.imshow(laplacian, cmap='gray', vmin=0, vmax=255)
ax2.title.set_text('Laplacian Filter')
```

# Sobel Edge Detection


```python
#Horizontal Edge Detection
sobel_h = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

# Apply kernel to image using cv2.filter2D
horizontal_edges = cv2.filter2D(image, -1, sobel_h).astype(np.float32)
#Vertical Edge Detection
sobel_v = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# Apply kernel to image using cv2.filter2D
vertical_edges = cv2.filter2D(image, -1, sobel_v).astype(np.float32)
#Both in Vertical and Horizontal Direction
sobel = cv2.magnitude(vertical_edges, horizontal_edges)

fig = plt.figure(figsize=(16,12))
ax1 = fig.add_subplot(141)
ax2 = fig.add_subplot(142)
ax3 = fig.add_subplot(143)
ax4 = fig.add_subplot(144)
ax1.imshow(image, cmap='gray', vmin=0, vmax=255)
ax1.title.set_text('Original Image')
ax2.imshow(horizontal_edges, cmap='gray', vmin=0, vmax=255)
ax2.title.set_text('Edges in Horizontal Direction')
ax3.imshow(vertical_edges, cmap='gray', vmin=0, vmax=255)
ax3.title.set_text('Edges in Vertical Direction')
ax4.imshow(sobel, cmap='gray', vmin=0, vmax=255)
ax4.title.set_text('Sobel Filtering')
```

# Prewitt Edge Detection


```python
#Horizontal Edge Detection
prewitt_h = np.array([[-1, -1, -1],
                      [0, 0, 0],
                      [1, 1, 1]])

# Apply kernel to image using cv2.filter2D
horizontal_edges = cv2.filter2D(image, -1, prewitt_h).astype(np.float32)
#Vertical Edge Detection
prewitt_v = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])

# Apply kernel to image using cv2.filter2D
vertical_edges = cv2.filter2D(image, -1, prewitt_v).astype(np.float32)
#Both in Vertical and Horizontal Direction
sobel = cv2.magnitude(vertical_edges, horizontal_edges)

fig = plt.figure(figsize=(16,12))
ax1 = fig.add_subplot(141)
ax2 = fig.add_subplot(142)
ax3 = fig.add_subplot(143)
ax4 = fig.add_subplot(144)
ax1.imshow(image, cmap='gray', vmin=0, vmax=255)
ax1.title.set_text('Original Image')
ax2.imshow(horizontal_edges, cmap='gray', vmin=0, vmax=255)
ax2.title.set_text('Edges in Horizontal Direction')
ax3.imshow(vertical_edges, cmap='gray', vmin=0, vmax=255)
ax3.title.set_text('Edges in Vertical Direction')
ax4.imshow(sobel, cmap='gray', vmin=0, vmax=255)
ax4.title.set_text('Prewitt Filtering')
```

**Canny Edge Detection**


- cv2.Canny: This function from the OpenCV library performs the Canny Edge Detection algorithm.

- First Argument (image): The input image on which edge detection is to be performed. It should be a single-channel (grayscale) image.

- Second Argument (THRESHOLD1): The lower threshold for hysteresis.

- Third Argument (THRESHOLD2): The upper threshold for hysteresis.

- Output (canny): The result is a binary image (same dimensions as the input image) where edges are marked with white pixels (255) and non-edges with black pixels (0).

# Canny Edge Detection Steps:

1. Noise Reduction: Typically done using a Gaussian filter to smooth the image and reduce noise.
2. Gradient Calculation: Computes the intensity gradient of the image using operators like Sobel.
3. Non-Maximum Suppression: Thins out the edges to achieve a thin line representation.
4. Hysteresis Thresholding: Uses the two thresholds to detect and link edges.


```python
THRESHOLD1 = 100
THRESHOLD2 = 200
canny = cv2.Canny(image, THRESHOLD1, THRESHOLD2)
fig = plt.figure(figsize=(10, 4))
ax1 = fig.add_subplot(121)
ax3 = fig.add_subplot(122)
ax1.imshow(image, cmap='gray', vmin=0, vmax=255)
ax1.title.set_text('Original Image')
ax3.imshow(canny, cmap='gray', vmin=0, vmax=255)
ax3.title.set_text('Canny Edge Detection')
```
